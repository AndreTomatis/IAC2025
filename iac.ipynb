```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arashnic/soalr-wind")

print("Path to dataset files:", path)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('./2/solar_wind.csv')

# Inspect the first few rows
print(df.head())
```

```python
df.drop(columns=['theta_gse', 'phi_gse', 'bx_gsm', 'by_gsm', 'bz_gsm', 'theta_gsm', 'phi_gsm'], inplace=True)
df.head()
```

```python
# Load the dataset
df_sunspot = pd.read_csv('./2/labels.csv')

# Inspect the first few rows
print(df_sunspot.head())
```

```python
df_combined = pd.concat([df.reset_index(drop=True), df_sunspot[['dst']].reset_index(drop=True)], axis=1)
df_combined.head()
```

```python
# Load the dataset
df_sat = pd.read_csv('./2/satellite_pos.csv')

# Inspect the first few rows
print(df_sat.head())
```

```python
import pandas as pd

# Convert 'timedelta' to actual timedelta objects
df_combined['timedelta'] = pd.to_timedelta(df_combined['timedelta'])
df_sat['timedelta'] = pd.to_timedelta(df_sat['timedelta'])

# Extract day as integer
df_combined['day'] = df_combined['timedelta'].dt.days
df_sat['day'] = df_sat['timedelta'].dt.days

# Keep only necessary columns from df_sat
satellite_columns = [
    'period', 'day',
    'gse_x_ace', 'gse_y_ace', 'gse_z_ace',
    'gse_x_dscovr', 'gse_y_dscovr', 'gse_z_dscovr'
]
df_sat_trimmed = df_sat[satellite_columns]

# Merge on 'period' and 'day'
df_merged = df_combined.merge(df_sat_trimmed, on=['period', 'day'], how='left')

# Create unified gse_x/y/z columns based on source
df_merged['gse_x'] = df_merged.apply(
    lambda row: row['gse_x_ace'] if row['source'] == 'ac' else row['gse_x_dscovr'], axis=1)
df_merged['gse_y'] = df_merged.apply(
    lambda row: row['gse_y_ace'] if row['source'] == 'ac' else row['gse_y_dscovr'], axis=1)
df_merged['gse_z'] = df_merged.apply(
    lambda row: row['gse_z_ace'] if row['source'] == 'ac' else row['gse_z_dscovr'], axis=1)

# Optionally drop the temporary and satellite-specific columns
df_final = df_merged.drop(columns=[
    'day',
    'gse_x_ace', 'gse_y_ace', 'gse_z_ace',
    'gse_x_dscovr', 'gse_y_dscovr', 'gse_z_dscovr'
])


df_final.head()
```
